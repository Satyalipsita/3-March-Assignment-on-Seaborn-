```python
#Q-1
Seaborn is a popular data visualization library in Python that provides 
a high-level interface for creating attractive statistical graphics.
Here are five commonly used plots in Seaborn along with their intended uses:

1. Scatter Plot:
   - Use: Displays the relationship between two numerical variables.
   - Syntax: `seaborn.scatterplot(x, y, data=None, **kwargs)`

2. Line Plot:
   - Use: Shows the trends or patterns in numerical data over a continuous interval.
   - Syntax: `seaborn.lineplot(x=None, y=None, data=None, **kwargs)`

3. Bar Plot:
   - Use: Compares the values of different categories using rectangular bars.
   - Syntax: `seaborn.barplot(x=None, y=None, data=None, **kwargs)`

4. Histogram:
   - Use: Represents the distribution of a single numerical variable.
   - Syntax: `seaborn.histplot(data=None, x=None, **kwargs)`

5. Heatmap:
   - Use: Displays the matrix-like data with color-encoded values.
   - Syntax: `seaborn.heatmap(data, cmap=None, **kwargs)`

These are just a few examples of the many types of plots that Seaborn offers. 
Each plot has its own unique purpose and can be customized further with additional
parameters to suit specific visualization needs.
```

```python
#Q-2:
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset('fmri')

# Create a line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='timepoint', y='signal', hue='region', style='event', data=fmri)
plt.title('FMRI Signal by Time Point for Different Regions and Events')
plt.xlabel('Time Point')
plt.ylabel('Signal')
plt.legend()
plt.show()
```

```python
#Q-3
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Create the first box plot (place vs age)
plt.figure(figsize=(10, 6))
sns.boxplot(x='embark_town', y='age', data=titanic)
plt.title('Age Distribution by Place')
plt.xlabel('Place')
plt.ylabel('Age')
plt.show()

# Create the second box plot (place vs fare)
plt.figure(figsize=(10, 6))
sns.boxplot(x='embark_town', y='fare', data=titanic)
plt.title('Fare Distribution by Place')
plt.xlabel('Place')
plt.ylabel('Fare')
plt.show()
```

```python
#Q-4
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds = sns.load_dataset('diamonds')

# Create a histogram with hue for cut
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds, x='price', hue='cut')
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Count')
plt.legend(title='Cut')
plt.show()

```

```python
#Q-6:
import seaborn as sns

# Load the flights dataset
flights = sns.load_dataset('flights')

# Reshape the data for heatmap
flights_data = flights.pivot('month', 'year', 'passengers')

# Create a heatmap
sns.heatmap(flights_data, cmap='YlGnBu')
```
